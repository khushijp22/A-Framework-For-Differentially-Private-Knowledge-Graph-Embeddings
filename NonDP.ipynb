```python
from pykeen.pipeline import pipeline
```

```python
import torch
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

#Run Knowledge Graph Embedding Pipeline


```python
# Step 3: Define your pipeline
result = pipeline(
    model='TransE',
    dataset='fb15k237',
    training_loop='slcwa',
    model_kwargs=dict(embedding_dim=50),
    optimizer='adam',
    optimizer_kwargs=dict(lr=0.01),
    training_kwargs=dict(num_epochs=100, batch_size=521),
    stopper='early',
    stopper_kwargs=dict(patience=5, frequency=5),
    random_seed=42,
    device='cuda',
)
```

#View Evaluation Metrics

```python
print("Mean Rank (MR):", result.get_metric("mean_rank"))
print("Mean Reciprocal Rank (MRR):", result.get_metric("mean_reciprocal_rank"))
print("Hits@1:", result.get_metric("hits_at_1"))
print("Hits@3:", result.get_metric("hits_at_3"))
print("Hits@10:", result.get_metric("hits_at_10"))
```
